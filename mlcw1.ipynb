```python
sort = regression_part2.sort_values(by = ['input'])

x = np.array(sort['input']).reshape(25,1)
y = np.array(sort['output']).reshape(25,1)


from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(1)
x1 = poly.fit_transform(x)

from sklearn.linear_model import LinearRegression
line1=LinearRegression(fit_intercept=False, normalize=False, copy_X=True, n_jobs=1)
line1.fit(x1,y)
plt.scatter(sort['input'],sort['output']) #画散点图看实际面积和价格的分布情况
plt.plot(sort['input'],line1.predict(x1),color='red',label = 'M=1')

poly = PolynomialFeatures(2)
x2 = poly.fit_transform(x)

from sklearn.linear_model import LinearRegression
line2=LinearRegression(fit_intercept=False, normalize=False, copy_X=True, n_jobs=1)
line2.fit(x2,y)
plt.plot(sort['input'],line2.predict(x2),color='y',label= 'M=2')

poly = PolynomialFeatures(3)
x3 = poly.fit_transform(x)
from sklearn.linear_model import LinearRegression
line3=LinearRegression(fit_intercept=False, normalize=False, copy_X=True, n_jobs=1)
line3.fit(x3,y)
plt.plot(sort['input'],line3.predict(x3),color='c',label= 'M=3')



poly = PolynomialFeatures(4)
x4 = poly.fit_transform(x)
from sklearn.linear_model import LinearRegression
line4=LinearRegression(fit_intercept=False, normalize=False, copy_X=True, n_jobs=1)
line4.fit(x4,y)
plt.plot(sort['input'],line4.predict(x4),color = 'b',label= 'M=4')

plt.xlabel('input', color = 'k')
plt.ylabel('output', color = 'k')
plt.title("different M for nonlinear regression", size=20)
plt.legend()
plt.show()



  
```

```python
mse1 =  mean_squared_error(y,line1.predict(x1) )
print(mse1)
mse2 = mean_squared_error(y,line2.predict(x2) )
print(mse2)
mse3 = mean_squared_error(y,line3.predict(x3) )
mse4 = mean_squared_error(y,line4.predict(x4) )
print(mse3)
print(mse4)
index = ('M1','M2','M3','M4')
values = (mse1, mse2, mse3, mse4)
width = 0.5 
p2 = plt.bar(index, values,width, color="#87CEFA")
plt.xlabel('four different polynomial regression models')
plt.ylabel('MSE') 
plt.title("bar plot for MSE of each of the four polynomial regression models", size=15)
```

```python
c1 = -4;
c2=-2.0;
c3=2.0;
c4=4.0;

a1 = 0.2;
a2=100;
a3 = 1000;

one = np.ones((25,1),dtype = int)

rbf11 = np.exp((-0.5*((x-c1)**2)) *(1/(a1**2)))
rbf12 =np.exp((-0.5*((x-c2)**2)) *(1/(a1**2)))
rbf13 =np.exp((-0.5*((x-c3)**2)) *(1/(a1**2)))
rbf14 =np.exp((-0.5*((x-c4)**2)) *(1/(a1**2)))
tmp1 = np.insert(rbf11,[0],one,axis=1)
tmp1 = np.insert(tmp1,[2],rbf12,axis=1)
tmp1 = np.insert(tmp1,[3],rbf13,axis=1)
tmp1 = np.insert(tmp1,[4],rbf14,axis=1)
d1=LinearRegression(fit_intercept=False, normalize=False, copy_X=True, n_jobs=1)
d1.fit(tmp1,y)
plt.scatter(sort['input'],sort['output']) #画散点图看实际面积和价格的分布情况
plt.plot(sort['input'],d1.predict(tmp1),color='red',label ='a=0.2')

rbf21 = np.exp((-0.5*((x-c1)**2)) *(1/(a2**2)))
rbf22 =np.exp((-0.5*((x-c2)**2)) *(1/(a2**2)))
rbf23 =np.exp((-0.5*((x-c3)**2)) *(1/(a2**2)))
rbf24 =np.exp((-0.5*((x-c4)**2)) *(1/(a2**2)))
tmp2 = np.insert(rbf21,[0],one,axis=1)
tmp2 = np.insert(tmp2,[2],rbf22,axis=1)
tmp2 = np.insert(tmp2,[3],rbf23,axis=1)
tmp2 = np.insert(tmp2,[4],rbf24,axis=1)
d2=LinearRegression(fit_intercept=False, normalize=False, copy_X=True, n_jobs=1)
d2.fit(tmp2,y)
plt.plot(sort['input'],d2.predict(tmp2),color='k',label ='a=100')

rbf31 = np.exp((-0.5*((x-c1)**2)) *(1/(a3**2)))
rbf32 =np.exp((-0.5*((x-c2)**2)) *(1/(a3**2)))
rbf33 =np.exp((-0.5*((x-c3)**2)) *(1/(a3**2)))
rbf34 =np.exp((-0.5*((x-c4)**2)) *(1/(a3**2)))
tmp3 = np.insert(rbf31,[0],one,axis=1)
tmp3 = np.insert(tmp3,[2],rbf32,axis=1)
tmp3 = np.insert(tmp3,[3],rbf33,axis=1)
tmp3 = np.insert(tmp3,[4],rbf34,axis=1)
d3=LinearRegression(fit_intercept=False, normalize=False, copy_X=True, n_jobs=1)
d3.fit(tmp3,y)
plt.plot(sort['input'],d3.predict(tmp3),color='b',label ='a=1000')
plt.xlabel('input', color = 'k')
plt.ylabel('output', color = 'k')
plt.title("different α for nonlinear regression", size=20)
plt.legend()
plt.show()
```

```python
face_train_path = os.path.join(os.getcwd(), 'data', 'faces_train_data.csv')
face_train = pd.read_csv(face_train_path)
print(face_train)
face_train.info()
face_train.size
```

```python
face_test_path = os.path.join(os.getcwd(), 'data', 'faces_test_data.csv')
face_test = pd.read_csv(face_test_path)
face_test.info()
```

```python
target_train = face_train['smiling']
smile_index = np.argwhere(target_train == 1)
nosmile_index = np.argwhere(target_train == 0)
print(nosmile_index)
print(smile_index)
```

```python
face_train1 = face_train[face_train['smiling'] == 1]
face_train1 = face_train1.drop('smiling',1)
face_mean1 = face_train1.mean(axis=0, skipna=False)
face_train2 = face_train[face_train['smiling'] == 0]
face_train2 = face_train2.drop('smiling',1)
print(face_train2)
face_mean2 = face_train2.mean(axis=0, skipna=False)
```

```python
dfx1=face_mean1[[i%2==0 for i in range(len(face_mean1.index))]]
dfy1=face_mean1[[i%2==1 for i in range(len(face_mean1.index))]]
dfx2=face_mean2[[i%2==0 for i in range(len(face_mean2.index))]]
dfy2=face_mean2[[i%2==1 for i in range(len(face_mean2.index))]]
plt.figure(figsize=(5,5)) 
plt.scatter(dfx1,dfy1,color = 'r',label ='smiling') 
plt.scatter(dfx2,dfy2,color = 'b',label ='nosmiling')
plt.xlabel('average of x')
plt.ylabel('average of y')
plt.title("two case of face", size=20)
plt.legend()
```

```python
face_trainx = face_train.drop('smiling',1)
face_trainy = face_train['smiling']
print(face_trainx)
print(face_trainy)
face_testx = face_test.drop('smiling',1)
face_testy = face_test['smiling']
print(face_testx)
print(face_testy)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
max_depth_list = [2,8,20]

dtc1 = DecisionTreeClassifier(random_state = 2001,max_depth=2) 
dtc1.fit(face_trainx,face_trainy)
train1_z = dtc1.predict(face_trainx)
test1_z = dtc1.predict(face_testx)
print("trainning Accuracy for max depth=2: " + str(accuracy_score(face_trainy, train1_z)*100)+"%")
print("test Accuracy for max depth=2: " + str(accuracy_score(face_testy, test1_z)*100)+"%")

dtc2 = DecisionTreeClassifier(random_state = 2001,max_depth=8) 
dtc2.fit(face_trainx,face_trainy)
train2_z = dtc2.predict(face_trainx)
test2_z = dtc2.predict(face_testx)
print("trainning Accuracy for max depth=8: " + str(accuracy_score(face_trainy, train2_z)*100)+"%")
print("test Accuracy for max depth=8: " + str(accuracy_score(face_testy, test2_z)*100)+"%")

dtc3 = DecisionTreeClassifier(random_state = 2001,max_depth=20) 
dtc3.fit(face_trainx,face_trainy)
train3_z = dtc3.predict(face_trainx)
test3_z = dtc3.predict(face_testx)
print("trainning Accuracy for max depth=20: " + str(accuracy_score(face_trainy, train3_z)*100)+"%")
print("test Accuracy for max depth=20: " + str(accuracy_score(face_testy, test3_z)*100)+"%")



```
